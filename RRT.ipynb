```python
def distancia(p1,p2):
    return sqrt((p1[0]-p2[0])*(p1[0]-p2[0])+(p1[1]-p2[1])*(p1[1]-p2[1]))

def step_from_to(nn,rand):
    if distancia(nn,rand) <= EPSILON:
        return rand
    else:
        theta = atan2(rand[1]-nn[1],rand[0]-nn[0])##Calculo de angulo
        aumento = nn[0] + EPSILON*cos(theta), nn[1] + EPSILON*sin(theta)
        return aumento
    
def obs(newnode):
    if((340>newnode[0]>260) and (280>newnode[1]>200)):
        return False
    elif((120>newnode[0]>80) and (440>newnode[1]>400)):
        return False
    else:
        return True
    
def obsPru(newnode):
    if((180>newnode[0]>0) and (480>newnode[1]>360)):
        return False
    elif((180>newnode[0]>0) and (120>newnode[1]>0)):
        return False
    elif((360>newnode[0]>180) and (220>newnode[1]>0)):
        return False
    elif((360>newnode[0]>180) and (480>newnode[1]>260)):
        return False
    else:
        return True
    
    
def recta(pu1,pu2):
    if(pu1[0] != pu2[0]):
        m = ((pu2[1]-pu1[1])/(pu2[0]-pu1[0])) 
    else:
        m = 0;
    obst = 0
    x = int(pu1[0])
    b = pu1[1] - m*pu1[0]
    y = m*x + b
    punto = x,y
    x += 1
    while(obsPru(punto) and(x < pu2[0])):
        y = m*x + b
        punto = x,y
        x += 1
    if((x == (int(pu2[0]))+1) or (x == 640)):
        return True
    else:
        return False
```

```python
def Dibujar():    
    pygame.init()    
    global screen
    screen = pygame.display.set_mode(WINSIZE)
    pygame.display.set_caption('RRT')    
    screen.fill(black)
    
    pygame.draw.rect(screen, verde, [0, 360, 180, 120], 0)
    pygame.draw.rect(screen, verde, [0, 0, 180, 120], 0)
    pygame.draw.rect(screen, verde, [180, 0, 180, 220], 0)
    pygame.draw.rect(screen, verde, [180, 260, 180, 220], 0)#"""
    
def NodoMasCercano(Arbol,q):
    nodo = Arbol[0]        
    for i in range(len(Arbol)):
        if (distancia(Arbol[i],q) < distancia(nodo,q)):
                nodo = Arbol[i]
    return nodo

def NuevoNodo(q_near,q):
    return step_from_to(q_near,q) 

def Choque(q,Arbol):
    cruzo = False
    for l in range(0,len(Arbol)-2):
        if((distancia(q,Arbol[l]) < EPSILON + 0.2)):
            cruzo = True
            Arbol[l+1] = q
            break
    return cruzo  
        
def Extiende(Arbol,q,aux):    
    global q_new
    global origen
    q_near = NodoMasCercano(Arbol,q)
    q_new = NuevoNodo(q_near,q)    
    if(obsPru(q_new)):
        Arbol.append(q_new)
        pygame.draw.line(screen,white,q_near,q_new)
        pygame.display.update()
        if(Choque(q_new,aux)):
            origen = q_new
            return 'Final'
        elif(q_new == q_near):
                return 'NO'
        else:
            return 'SI'
    else:
        return 'NO'
    

def Conecta(Arbol,q,aux):
    s = Extiende(Arbol,q,aux)
    while(s == 'SI'):
        s = Extiende(Arbol,q,aux)       
    #print("v->",s)
    return s

def Intercambiar(arbolA, arbolB):
    return arbolB,arbolA

def Camino(union,Arbol):
    busca = True
    ruta = []
    while(busca):
        for v in range(0,len(ruta)):
            ruta.pop()
        ruta.append(Arbol[0])
        r = 1        
        u = ruta[0]                        
        for m in range(1,len(Arbol)):
            s = Arbol[m]        
            if(distancia(u,s) < EPSILON + 0.2):
                ruta.append(Arbol[m])
                u = ruta[r]
                r += 1
                a = m
        if((ruta[r-1]) == union):
            busca = False
        else:
            Arbol[a] = 0,0 
            
    return ruta

def Mini(Arbol):
    final = []
    NumVer = len(Arbol)-1
    cosas = 1
    pru = 0
    #NumVer = int(NumVer/2)
    v1 = Arbol[0]   
    final.append(v1)
    v2 = Arbol[NumVer]
    while(final[len(final)-1] != Arbol[len(Arbol)-1]):
        pygame.draw.rect(screen, otro, [int(v2[0]),int(v2[1]),10,10])
        pygame.display.update()
        sleep(0.1)   
        if((distancia(v1,v2) < EPSILON + 0.2) or (recta(v1,v2))):
            final.append(v2)
            for ve in range (NumVer,len(Arbol)):
                final.append(Arbol[ve])              
            v1 = Arbol[NumVer]
            Arbol = final.copy()
            NumVer = len(Arbol)-1
            v2 = Arbol[NumVer]
            cosas += 1
            for bo in range (cosas,len(final)):
                final.pop()            
        else:
            if(NumVer > 4):
                NumVer = int(8*NumVer/9)
            else:
                NumVer -= 1
            v2 = Arbol[NumVer]
            
    return final
    
```
